```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'city': ['City A', 'City B', 'City A', 'City C', 'City B', 'City C'],
    'complaint_type': ['Type A', 'Type B', 'Type A', 'Type C', 'Type B', 'Type C']
}

df = pd.DataFrame(data)

# Perform count in group by
grouped_counts = df.groupby(['city', 'complaint_type']).size().reset_index(name='count')

# Create a bar plot
plt.bar(grouped_counts['city'], grouped_counts['count'])

# Set x-axis label
plt.xlabel('City')

# Set y-axis label
plt.ylabel('Count')

# Set plot title
plt.title('Counts of Complaint Types by City')

# Display the plot
plt.show()
```
